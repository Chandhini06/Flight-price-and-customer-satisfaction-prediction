```python
pd.options.mode.chained_assignment = None

df["Date_of_Journey"] = pd.to_datetime(df["Date_of_Journey"])

df['Year'] = df["Date_of_Journey"].dt.year
```

```python
pd.options.mode.chained_assignment = None

df["Date_of_Journey"] = pd.to_datetime(df["Date_of_Journey"])

df['Month'] = df["Date_of_Journey"].dt.month
```

```python
pd.options.mode.chained_assignment = None

df["Date_of_Journey"] = pd.to_datetime(df["Date_of_Journey"])

df['Day'] = df["Date_of_Journey"].dt.day
```

```python
df
```

```python
df.drop(["Date_of_Journey"], axis = 1,  inplace = True)
```

```python
df
```

```python
df["Duration_Hours"] = df["Duration"].str.extract(r'(\d+)h').fillna(0).astype(int)
df["Duration_Min"] = df["Duration"].str.extract(r'(\d+)m').fillna(0).astype(int)

df.drop(["Duration"], axis = 1, inplace = True)
```

```python
df
```

```python
df["Dep_Hour"] =  df["Dep_Time"].astype(str).str.split(":").str[0].astype(int)
df["Dep_Min"] =  df["Dep_Time"].astype(str).str.split(":").str[1].astype(int)

df.drop(["Dep_Time"], axis = 1, inplace = True)

df
```

```python
df["Arrival_Time"] = df["Arrival_Time"].astype(str)

df["Extracted_Time"] = df["Arrival_Time"].str.extract(r"(\d{1,2}:\d{2})")


df["Arrival_Hour"] = df["Extracted_Time"].str.split(":").str[0].astype(int)
df["Arrival_Min"] = df["Extracted_Time"].str.split(":").str[1].astype(int)

df = df.drop(columns = ["Arrival_Time"])
df
```

```python
df = df.drop(["Extracted_Time"], axis = 1)
df
```

```python
df['Total_Stops'] = df['Total_Stops'].replace({
    "non-stop": 0,
    "1 stop": 1,
    "2 stops": 2,
    "3 stops": 3,
    "4 stops": 4
})

df
```

```python
# Label Encoding 

from sklearn.preprocessing import LabelEncoder

model = LabelEncoder()
df['Route'] = model.fit_transform(df['Route'])

df
```

```python
# One-Hot Encoding 

df = pd.get_dummies(df, columns = ['Airline', 'Source', 'Destination', 'Additional_Info'], dtype = 'int')
df
```

```python
df['Price_per_Min'] = df['Price']/((df['Duration_Hours']*60) + df['Duration_Min'])

df
```

# ML

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, accuracy_score
```

```python
X = df.drop(['Price'], axis = 1)
y = df['Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model_params = {
    "Linear Regression": {
        "model": LinearRegression(),
        "params": {}
    },
    "Random Forest": {
        "model": RandomForestRegressor(random_state=42),
        "params": {"n_estimators": [150], "max_depth": [2, 3, 4, 5, 6, 7, 10, 15], "min_samples_split": [10]}
    },
    "Gradient Boosting": {
        "model": GradientBoostingRegressor(random_state=42),
        "params": {"n_estimators": [150], "learning_rate":[0.1,0.01]}
    },
    "Decision Tree": {
        "model": DecisionTreeRegressor(random_state=42),
        "params": {"max_depth": [2, 3, 4, 5, 6, 7, 10, 15], "min_samples_split": [10]}
    },
    "XGBoost": {
        "model": XGBRegressor(random_state=42, objective="reg:squarederror"),
        "params": {"n_estimators": [150], "learning_rate": [0.1, 0.01]}
    }
}



for name, mp in model_params.items():
    print(f"Training and Tuning Model: {name}")

    # Perform GridSearchCV only if there are parameters to tune
    if mp["params"]:
        grid_search = GridSearchCV(mp["model"], mp["params"], cv=5, scoring="neg_mean_squared_error", n_jobs=-1)
        trained_model = grid_search.fit(X_train, y_train)
        best_model = grid_search.best_estimator_
        print(f"Best Parameters: {grid_search.best_params_}")
    else:
        best_model = mp["model"]
        best_model.fit(X_train, y_train)

    # Predictions
    train_pred = best_model.predict(X_train)
    test_pred = best_model.predict(X_test)


   
    print("******* Train *******")
    print(f"MSE: {mean_squared_error(y_train, train_pred)}")
    print(f"MAE: {mean_absolute_error(y_train, train_pred)}")
    print(f"R2 Score: {r2_score(y_train, train_pred)}")

    print("******* Test *******")
    print(f"MSE: {mean_squared_error(y_test, test_pred)}")
    print(f"MAE: {mean_absolute_error(y_test, test_pred)}")
    print(f"R2 Score: {r2_score(y_test, test_pred)}")
    print("\n")

```

```python
import pickle
filename = f"{name.replace(' ', '_').lower()}_model.pkl"  
with open(filename, "wb") as f:
    pickle.dump(best_model, f)

    print(f" Saved {name} model as {filename}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
pd.read_csv("C:/Users/Admin/OneDrive/Documents/Flight ML project/samples.csv")
```

```python
X = samples.drop("satisfaction", axis=1)
y = samples["satisfaction"]
```

```python
print(samples["satisfaction"].value_counts())
```

```python
from imblearn.under_sampling import RandomUnderSampler

rosee = RandomUnderSampler(random_state=42)
X_resampled, y_resampled = rosee.fit_resample(X, y)

print(pd.Series(y_resampled).value_counts())
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
```

```python
models = [LogisticRegression(random_state=42), XGBClassifier(random_state=42), DecisionTreeClassifier(max_depth=10, min_samples_split=20, min_samples_leaf=10,random_state=42),  RandomForestClassifier(n_estimators=50, max_depth=15, random_state=42, ), GradientBoostingClassifier(random_state=42)]

for model in models:
    
    model.fit(X_train, y_train)
    train_pred = model. predict(X_train)
    test_pred = model. predict(X_test)
    
    print (f"Model: {type(model).__name__}")
    
    print("*******Train*******")
    print (f"Accuracy Score: {accuracy_score(y_train, train_pred)}")
    print(f"Precision Score: {precision_score(y_train,train_pred)}")
    print(f"Recall score: {recall_score(y_train, train_pred )}")
    print(f"F1 Score: {f1_score(y_train, train_pred)}")
    print ("\n")
    
    print ("********Test********")
    print (f"Accuracy Score: {accuracy_score(y_test, test_pred) }")
    print(f"Precision Score: {precision_score(y_test, test_pred )}")
    print(f"Recall score: {recall_score(y_test, test_pred)}")
    print(f"F1 Score: {f1_score(y_test, test_pred )}")

    print(f"\nConfusion Matrix for {model}:\n", confusion_matrix(y_test, test_pred))
    print ("\n")
```

```python
import pickle

best_model = None
best_model_name = ""
best_accuracy = 0  

# Train, Evaluate, and Compare Models
for model in models:
    model.fit(X_train, y_train)
    test_pred = model.predict(X_test)

    model_name = model.__class__.__name__  
    accuracy = accuracy_score(y_test, test_pred)  

    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_model = model
        best_model_name = model_name

# Save the Best Model
if best_model:
    filename = f"best_model_{best_model_name.lower()}.pkl"
    with open(filename, "wb") as f:
        pickle.dump(best_model, f)
    print(f"\n Best Model: {best_model_name} saved as {filename} with Accuracy: {best_accuracy:.4f}")
```
